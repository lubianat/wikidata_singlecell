```python
data =data[["MTAB" in string for string in data["Data location"]]]
len(data)
```

```python
def explode_commas(df, column):

    strings = []
    for string in df[column]:
        
        if "," in string:
            array = string.split(", ")
            strings.append(array)
        else:
            strings.append(string)

    df[column] = strings
    print(df)
    return(df.explode(column))
```

Let us explode those commas

```python
exploded_data = explode_commas(data, "Data location")

print(len(data))
print(len(exploded_data))
```

```python
def make_mtab_url(mtab):
    return('"' + "https://www.ebi.ac.uk/arrayexpress/experiments/" + mtab + '"')
```

```python
from time import gmtime, strftime


with open("external_data_11_02_2021.qs", "w") as f:
    for i, row in exploded_data.iterrows():
        s = row["item"]
        p = "P1325" # external data available at
        o = make_mtab_url(row["Data location"])
        rp1 = "S248"
        ro1 = "Q103034964"
        rp2 = "S854"
        ro2 = '"https://www.nxn.se/single-cell-studies/data.tsv"'
        rp4 = "S813"
        ro4 = strftime("+%Y-%m-%dT00:00:00Z/11", gmtime())
        print(ro4)
        f.write(f"{s}|{p}|{o}|{rp1}|{ro1}|{rp2}|{ro2}|{rp4}|{ro4}")
        f.write("\n")
```
